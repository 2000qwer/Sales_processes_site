```python
import pymongo
from flask import Flask, jsonify, json  # Import ObjectId from bson module
from urllib.parse import quote_plus
import logging
from flask_cors import CORS


app = Flask(__name__)
CORS(app)
logging.basicConfig(level=logging.DEBUG)


# MongoDB connection setup
username = "MichalDataBase"
password = "Qwer1997@"
cluster = "cluster0.o2qdros.mongodb.net"

# URL-encode the username and password
escaped_username = quote_plus(username)
escaped_password = quote_plus(password)

# Build the MongoDB connection string
connection_string = f"mongodb+srv://{escaped_username}:{escaped_password}@{cluster}/test?retryWrites=true&w=majority"

# Create a MongoClient to connect to the MongoDB cluster
client = pymongo.MongoClient(connection_string)

# Access your database and collection
db = client.test
collection_posts = db.posts
collection_users = db.users
collection_profiles = db.profiles
collection_sales = db.sales

# Define routes



@app.route('/users')
def get_users_from_mongodb():
    # Query data from MongoDB
    users_from_mongodb = list(collection_users.find())

    # Convert ObjectId to string representation
    serialized_users = json.dumps(users_from_mongodb, default=str)

    # Convert JSON string to response object
    json_users = jsonify(json.loads(serialized_users))

    return json_users


@app.route('/sales')
def get_sales_from_mongodb():
    # Query data from MongoDB
    sales_from_mongodb = list(collection_sales.find())

    # Convert ObjectId to string representation
    serialized_sales = json.dumps(sales_from_mongodb, default=str)

    # Convert JSON string to response object
    json_sales = jsonify(json.loads(serialized_sales))

    return json_sales


@app.route('/lastsale')
def get_last_sale_from_mongodb():
    # Retrieve the latest document in the collection
    latest_sale = collection_sales.find().sort('_id', -1).limit(1)

    # Convert ObjectId to string representation
    serialized_last_sale = json.dumps(latest_sale[0], default=str)

    # Convert JSON string to response object
    json_last_sale = jsonify(json.loads(serialized_last_sale))

    return json_last_sale 



@app.route('/profiles')
def get_profiles_from_mongodb():
    # Query data from MongoDB
    profiles_from_mongodb = list(collection_profiles.find())

    # Convert ObjectId to string representation
    serialized_profiles = json.dumps(profiles_from_mongodb, default=str)

    # Convert JSON string to response object
    json_profiles = jsonify(json.loads(serialized_profiles))

    return json_profiles


@app.route('/posts')
def get_posts_from_mongodb():
    # Query data from MongoDB
    posts_from_mongodb = list(collection_posts.find())

    # Convert ObjectId to string representation
    serialized_posts = json.dumps(posts_from_mongodb, default=str)

    # Convert JSON string to response object
    json_posts = jsonify(json.loads(serialized_posts))

    return json_posts


if __name__ == '__main__':
    # Use 0.0.0.0 to listen on all public IPs
    app.run(host='0.0.0.0', port=8080, debug=False)
```
