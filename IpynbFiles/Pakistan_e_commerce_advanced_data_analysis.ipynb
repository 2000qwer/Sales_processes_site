```python
def make_autopct(values):
    def my_autopct(pct):
        total = sum(values)
        val = int(round(pct*total/100.0))
        return '{p:.2f}%  ({v:d})'.format(p=pct,v=val)
    return my_autopct


for i in df_distinct.itertuples():
    plt.figure(figsize=(15, 13))
    palette_color = sns.color_palette('dark')
    plt.pie(df_c_p['Customer count buy'].loc[df_c_p['Category Name'] == str(i[1])], labels=df_c_p['payment_method'].loc[df_c_p['Category Name'] == str(i[1])], autopct='%1.1f%%', startangle=140)
    plt.title(str(i[1]))
    plt.show()
```
